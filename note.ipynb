```python
import numpy as np
import matplotlib.pyplot as plt

# Correct range using numpy
x = np.arange(0.01, 1, 0.01)  # Avoid log(0), which is undefined
y = np.log2(x)  # Use numpy.log2() for element-wise calculation

# Plot
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("log2(x)")
plt.title("Plot of log2(x)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
A = 1           # Amplitude
P = 5      # Custom pitch (one full cycle spans this value)
phi = 0        # Phase offset
t = np.linspace(0, 2*P, 1000)  # Time (or spatial) domain

# Generate sine wave
y = A * np.sin(2 * np.pi * t / P + phi)

# Plot the sine wave
plt.plot(t, y)
plt.axvline(x=P, color='r', linestyle='--', label="One Cycle")
plt.title(f"Sine Wave with One Cycle Spanning {P} Units")
plt.xlabel("Time (or Spatial Position)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```
